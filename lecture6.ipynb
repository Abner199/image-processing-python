## 1. Digital Image (数字图像)
[Top](#sec00) | [Previous: Content](#sec01) |  [Next: Basic Operations](#sec03) 

- A digital image is a **numeric representation**, normally **binary**, of a two-dimensional image.   
- Depending on whether the image resolution is fixed, it may be of vector or raster type.   
- By itself, the term "digital image" usually refers to **raster images** or **bitmapped images** (as opposed to vector images).
- Common image formats: **PPM, PNG, JPEG, GIF, TIFF**, and **BMP**.   

### Gray image

- For monochrome (gray) images, the brightness of **each pixel** is expressed by a **numerical value**, 
- usually in the range of **0 to 255**, **0 for black**, **255 for white**, and other values for gray between black and white.

### Color image

- Color images can be represented by **two-dimensional matrices** of **Red**, **Green** and **Blue** triples.  
  - Did you hear about **RGB** ?  
- Normally, each value of a triple is between 0 and 255, 
  - 0 means that the corresponding primary color is not present in the pixel, 
  - and 255 means that the corresponding primary color is maximized in the pixel.

## 3. Transformation
[Top](#sec00) | [Previous: 2. Basic Operations](#sec03)  |  [Next: 4. Conversion](#sec05) 

This part will tell you basic image transforms such as **Crop, Rotate, Resize** an image, etc.

### Crop an image  

If your image is too big or some parts of the image is useful, you can crop it.

Here is an example.


```python
## 
from PIL import Image  
import matplotlib.pyplot as plt

## 
img=Image.open('./street-sign.jpg')

## 
region = (100,100,700,450)
cropImg = img.crop(region)

## 
plt.imshow(cropImg)
plt.show()

# The result will be showed below. 
```

### Rotate an image

If some images are not angeled properly, you may need to rotate them to make them look better.

Here is an example.

```python
## ROTATE_90 means rotate 90 degree
img_rotate = img.transpose(Image.ROTATE_90) 

##  
plt.imshow(img_rotate)
plt.show()

# The result will be showed below. 
```

## 5. Binarization
[Top](#sec00) | [Previous: 4. Conversion](#sec05)  |  [Next: 6. Common Filtering ](#sec07)

- Image binarization is the process of setting the gray value of the pixels on the image to **0 or 255**, 
  - that is to say, the whole image presents an obvious **black-and-white** effect.

- In digital image processing, binary image plays a very important role. 
  - The binarization of an image greatly **reduces the amount of data** in the image, so that the outline of the object can be highlighted.

- In Pillow, we use **point()** method to judge each pixel of an image, and set its gray value to 0 or 255 according to its relative size with the mean value, before converting it into a binary image.



```python
WHITE, BLACK = 255, 0   # define gray constant value

img_point = img.point(lambda x: WHITE if x > 128 else BLACK)  # judge each pixel

img_binary = img_point.convert('1')  # convert to binary image

## 
plt.imshow(img_binary)
plt.show()

# The result will be showed below. 
```

## 6. Common Filtering
[Top](#sec00) | [Previous: 5. Binarization](#sec06)  |  [Next: 7. Advanced Filtering ](#sec08)

- Image filtering in Python mainly deals with image **Smoothing, Sharpening** and **Edge enhancement**. 
  - You can regard it as a filter in a mobile photo processing app, like *美图秀秀*.

- There are many kinds of filtering in Pillow, like **Relief Filtering**.


```python
## 
from PIL import Image,ImageFilter  

## EMBOSS means filter type we used.
img_filter = img.filter(ImageFilter.EMBOSS) 

## TODO: you can choose anther method based on your work.

## 
plt.imshow(img_filter)
plt.show()

# The result will be showed below. 
```

We can now change to other filters such as EDGE_ENHANCE to find the egdes of objects in the image.

```python
##
from PIL import Image,ImageFilter  

## EDGE_ENHANCE filter, to find traffic sign edge better.
img_filter = img.filter(ImageFilter.EDGE_ENHANCE) 

## 
plt.imshow(img_filter)
plt.show()

# The result will be showed below. 
```

We can now change to other filters such as CONTOUR to find the contours of objects in the image.

```python
##
from PIL import Image,ImageFilter  

## CONTOUR filter, to find traffic sign contour.
img_filter = img.filter(ImageFilter.CONTOUR)  

## 
plt.imshow(img_filter)
plt.show()

# The result will be showed below. 
```

## 7. Advanced Filtering: Image Denosing
[Top](#sec00) | [Previous: 6. Common Filtering](#sec07)  |  [Next: 8. Histogram](#sec09)

Some images have some noise which makes it difficult to evaluate, we can use some filters to remove these noise.

Here we give an example using **Median Filter** to remove noise in the image.

- The Median Filter is a nonlinear digital filtering technique, often used to remove noise from an image or signal. 
- Such noise reduction is a typical **pre-processing** step to improve the results for later processing (for example, edge detection on an image). 
- Median filtering is widely used in digital image processing because under certain conditions it **preserves edges while removing noise** (see discussion below), also having applications in signal processing.

*Note: This image is "street-sign_noise.jpg", not "street-sign.jpg".*


```python
### read a image then display it###
from PIL import Image  
import matplotlib.pyplot as plt

## Open the image
img_noise=Image.open('./street-sign_noise.jpg')

## 
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img_noise)
plt.subplot(1,2,2)

## MedianFilter, to remove noise
img_filter = img_noise.filter(ImageFilter.MedianFilter(5)) 

## 
plt.imshow(img_filter)
plt.show()

# The result will be showed below. 
```

You can try other filters to remove noise such as a Gaussian filter, and also try to change different parameters in "img_noise.filter(ImageFilter.GaussianBlur(radius=2))" and check its results.

Try to undertand its principle and corresponding results.

```python
### read a image then display it###
from PIL import Image  
import matplotlib.pyplot as plt

## Open the image
img_noise=Image.open('./street-sign_noise.jpg')

##
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img_noise)
plt.subplot(1,2,2)

## GaussianBlur, to to remove noise.
img_filter = img_noise.filter(ImageFilter.GaussianBlur(radius=2)) 

##
plt.imshow(img_filter)
plt.show()

# The result will be showed below. 
```

## 8. Histogram (直方图)
[Top](#sec00) | [Previous: 7. Advanced Filtering](#sec08) 

An image histogram is a type of histogram that acts as **a graphical representation of the tonal distribution in a digital image**. It plots the number of pixels for each tonal value. By looking at the histogram for a specific image a viewer will be able to judge the entire tonal distribution at a glance

1. Histogram is a graphical representation of the intensity distribution of an image.
2. Histogram quantifies the number of pixels for each intensity value.

Here is an example of the histogram for color images.

```python
## 
from PIL import Image  
import matplotlib.pyplot as plt

## Use OpenCV here
import cv2 
img = cv2.imread('./street-sign.jpg')

## Convert RGB color image to gray scale image
img = cv2.cvtColor (img, cv2.COLOR_BGR2GRAY) 

## 
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img,cmap=plt.cm.gray )
plt.subplot(1,2,2)

## Histogram
plt.hist(img.ravel(),256,[0,256])

# The result will be showed below.
```
